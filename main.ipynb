# 基于深度学习的井下钻场图像视频分析系统

## 01. 项目背景

#### 1.1 煤炭产业概述

   * 我国是世界上最大的煤炭生产国和消费国，以煤为主的能源结构是我国能源资源赋存现状的必然选择，并将长期存在。因此，钻孔深度验收作为矿井作业的一项重要任务，需要行业投入大量人力物力。由于退钻作业事关矿下安全问题，每个矿山都会安排专门人员两班倒全天去监视钻杆退钻、进钻情况，钻孔深度与打入钻杆根数对应，当前煤矿采用的钻孔深度验收方法大多为人工验收法，由安全员在打钻现场监督打钻作业，或在监控室查看打钻录像，检查打钻工人是否按作业要求打够规定钻杆数。由于钻场管理人员难以连续监控现场视频，就可能出现工人施工谎报进尺、打假孔等行为，进而影响工程设计效果、引起安全事故。人工验收模式出现耗时耗力、效率低下、提供信息不够准确、不能实时提供人员状态信息和瓦斯抽采过程，且无法彻底避免人为因素给钻孔验收带来的不确定性等。当井下事故发生时，由于井下作业人员状况不明、监控视频模糊、通信方式简单，不能及时掌握井下人员动态信息，极易造成救援时间、救援资金浪费等现象。
   * 所以为解决以上问题本作品采用深度学习技术对煤矿井下人员进行实时检测，确保人员作业处于安全状态，同时对钻孔视频进行检测和跟踪，并对瓦斯抽采过程中钻杆进行计数，可以保障瓦斯抽采工作保时保量地完成，从而提升井下监管水平，加强煤矿安全管理。

#### 1.2 煤矿打钻作业的重要性及打钻作业的验收标准

   * 钻孔深度验收是打钻作业中极为重要的一项任务；煤矿打钻作业作为采矿作业前的重要探查工作，起到探放水、瓦斯抽采、掘取煤样等重要作用。
   * 合格的打钻作业是保证采矿作业安全的重要前提，钻孔深度作为打钻作业最重要的验收指标，一旦不达标将极有可能造成煤样瓦斯抽采含量低、探放水工作不到位、煤层抽采浓度低等问题，为后续开采工作埋下严重安全隐患，威胁井下矿工的生命安全。
   
#### 1.3 行业痛点

   * 当前煤矿采用的钻孔深度验收方法大多为人工验收法，由安全员在打钻现场监督打钻作业，或在监控室查看打钻录像，检查打钻工人是否按作业要求打够规定钻杆数。
   * 目前存在的问题：

#### 先导入必要的库

```python
import cv2
import paddle
import numpy as np
import pandas as pd
from paddle.vision.transforms import Compose, Resize, Normalize
from paddle.vision.models import yolov5
```

#### 实现第一步：对视频流进行处理，得到数据并保存

```python
# 定义模型路径和视频路径
model_path = 'best_model.pdema'
video_path = 'my.mp4'
output_csv = 'tuizuan.csv'

# 加载模型
model = paddle.jit.load(model_path)
model.eval()

# 定义预处理步骤
transform = Compose([
    Resize((640, 640)),
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# 打开视频文件
cap = cv2.VideoCapture(video_path)

# 创建一个空的 DataFrame 来保存坐标
coordinates_df = pd.DataFrame(columns=['frame', 'x_center', 'y_center'])

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 预处理图像
    input_tensor = transform(frame)
    input_tensor = paddle.to_tensor(input_tensor).unsqueeze(0)

    # 进行推理
    with paddle.no_grad():
        outputs = model(input_tensor)

    # 解析输出
    outputs = outputs.numpy()
    boxes = outputs[:, :4]
    scores = outputs[:, 4]
    classes = outputs[:, 5]

    # 提取检测框中心点坐标
    for box in boxes:
        x_min, y_min, x_max, y_max = box
        x_center = (x_min + x_max) / 2
        y_center = (y_min + y_max) / 2

        # 将坐标添加到 DataFrame
        coordinates_df = coordinates_df.append({
            'frame': frame_count,
            'x_center': x_center,
            'y_center': y_center
        }, ignore_index=True)

    frame_count += 1

# 保存坐标到 CSV 文件
coordinates_df.to_csv(output_csv, index=False)

# 释放视频捕获对象
cap.release()

print(f"检测框中心点坐标已保存到 {output_csv}")
```

#### 实现第二步：计数逻辑

```python
def calculate_drill_by_Distance(y_values):
    # 初始化一个空列表，用于存储钻孔位置（但在这个函数中并没有被使用）
    drill_positions = []
    
    # 将输入的 y_values 转换为 NumPy 数组，以便进行高效的数值计算
    values = np.array(y_values)
    
    # 计算相邻元素的差值，并取绝对值，得到一个包含这些差值的新数组
    diff_list = np.abs(np.diff(values))
    
    # 计算所有相邻元素差值的绝对值之和
    diff_sum = np.sum(diff_list)
    
    # 将差值总和除以 2，得到一个平均值(退钻动作为 推、拉 一个来回，所以除以二)
    diff_single = diff_sum / 2
    dill_count = diff_single // 407
    
    # 返回计算得到的钻孔数量
    return dill_count
```
